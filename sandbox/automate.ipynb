```python
def read_effect(left_ratio, top_ratio, right_ratio, bottom_ratio):
    screen_size = pyautogui.size()
    
    w_ratio = right_ratio - left_ratio
    h_ratio = bottom_ratio - top_ratio
    
    im = pyautogui.screenshot(region=(
        int(screen_size.width * left_ratio), int(screen_size.height * top_ratio),
        int(screen_size.width * w_ratio), int(screen_size.height * h_ratio)
    ))

    print(pytesseract.image_to_string(im).replace('\n', ' ').strip())

    return im
```

```python
read_effect(
    1476 / screen_size.width, 1077 / screen_size.height,
    2210 / screen_size.width, 1157 / screen_size.height
    )
```
